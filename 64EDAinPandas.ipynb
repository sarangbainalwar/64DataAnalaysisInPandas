```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numeric columns
numeric_df = df.select_dtypes(include=['number'])

# Create a heatmap
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm", fmt=".2f")

# Show the plot
plt.rcParams['figure.figsize'] = [10, 7]
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['number'])
print(numeric_df.corr())
```

```python
# Select numeric columns only
numeric_df = df.select_dtypes(include=['number'])

# Add the 'Continent' column back to the numeric DataFrame for grouping
numeric_df['Continent'] = df['Continent']

# Group by 'Continent' and calculate the mean
grouped_mean = numeric_df.groupby('Continent').mean().sort_values(by="2022 Population", ascending=False)

print(grouped_mean)
```

```python
numeric_df = df.select_dtypes(include=['number'])

# Add the 'Continent' column back to the numeric DataFrame for grouping
numeric_df['Continent'] = df['Continent']

# Group by 'Continent' and calculate the mean
grouped_mean = numeric_df.groupby('Continent')[['1970 Population', '1980 Population', '1990 Population',
       '2000 Population', '2010 Population', '2015 Population',
       '2020 Population', '2022 Population']].mean().sort_values(by="2022 Population", ascending=False)

print(grouped_mean)
```
